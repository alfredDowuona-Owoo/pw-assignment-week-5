###### Q1. Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions.

We use the Exception class as the base class when creating a custom exception because it ensures that our custom exception inherits all the necessary behavior and functionality of standard Python exceptions. By inheriting from the Exception class, our custom exception becomes compatible with the existing exception handling mechanisms, allowing us to catch and handle it like any other exception in Python

##### Q2. Write a python program to print Python Exception Hierarchy. please be brief with your answer

```python
import builtins

def print_exception_hierarchy(exception_cls, indent=0):
    print(' ' * indent + str(exception_cls))
    for subclass in exception_cls.__subclasses__():
        print_exception_hierarchy(subclass, indent + 4)

print_exception_hierarchy(builtins.BaseException)
```

The ArithmeticError class is a base class for arithmetic-related exceptions in Python. Two common errors defined in this class are:

ZeroDivisionError: Raised when attempting to divide by zero.

OverflowError: Raised when a numeric operation exceeds the bounds of the numeric type.

##### Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class in Python is used as a base class for exceptions related to sequence and mapping (dictionary) lookups. It provides a common base for exceptions like KeyError and IndexError, which are used to handle errors when trying to access elements in a sequence (e.g., lists, tuples, strings) or mappings (e.g., dictionaries) that do not exist.

KeyError: Raised when you try to access a dictionary key that does not exist.

```python
my_dict = {'apple': 1, 'banana': 2}
try:
    value = my_dict['cherry']  # This will raise a KeyError
except KeyError as e:
    print("KeyError:", e)
```

IndexError: Raised when you try to access an index in a sequence (e.g., list, tuple, string) that is out of range.

```python
my_list = [1, 2, 3]
try:
    value = my_list[4]  # This will raise an IndexError
except IndexError as e:
    print("IndexError:", e)
```

##### Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError is a Python exception that occurs when there is an issue with importing a module or a symbol (such as a function or variable) from a module.

ModuleNotFoundError is a specific subclass of ImportError that occurs when Python cannot locate the module you are trying to import.
